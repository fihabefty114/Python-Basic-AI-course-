```python
# 6B) Function to get maximum from a list
def max_in_list(items):
    if not items:
        return None
    m = items[0]
    for x in items[1:]:
        if x > m:
            m = x
    return m

data = [3, 1, 8, 2, 5]
print("Max:", max_in_list(data))
```

```python
# 6C) Count even numbers in a list
def count_evens(items):
    count = 0
    for x in items:
        if x % 2 == 0:
            count += 1
    return count

data = [1, 2, 3, 4, 6, 7, 8]
print("Even count:", count_evens(data))
```

```python
# 6D) Reverse a list without reverse()
def reverse_list(items):
    res = []
    for i in range(len(items)-1, -1, -1):
        res.append(items[i])
    return res

data = [10, 20, 30, 40]
print("Original:", data)
print("Reversed:", reverse_list(data))
```

```python
# 6E) Factorial using recursion
def factorial(n):
    if n < 0:
        raise ValueError("factorial undefined for negative numbers")
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

print("factorial(5) =", factorial(5))
```

```python
rows = 5
width = 2 * rows  

for i in range(1, rows):
    print(('*' * (2*i - 1)).center(width))

print(('*' * width).center(width))

for i in range(rows-1, 0, -1):
    print(('*' * (2*i - 1)).center(width))
```

```python
n = 5  
for i in range(n):
    line = ""
    for j in range(n):
        if i in (0, n-1) or j in (0, n-1):
            line += "*"
            
        else:
            line += "  "
    print(line)
```
