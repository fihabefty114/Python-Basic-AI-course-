```python
# 1) List Operations (dedupe + manual max/min/avg without max()/sum())
nums = [4, 7, 2, 7, 9, 2, 5, 9, 1, 4]
print("Original:", nums)

# Remove duplicates while preserving order
seen = set()
dedup = []
for x in nums:
    if x not in seen:
        seen.add(x)
        dedup.append(x)
print("Dedup (order kept):", dedup)

if not dedup:
    print("Empty list; no stats.")
else:
    mmax = dedup[0]
    mmin = dedup[0]
    total = 0
    count = 0
    for v in dedup:
        if v > mmax: mmax = v
        if v < mmin: mmin = v
        total += v
        count += 1
    avg = total / count
    print("Max:", mmax, "Min:", mmin, "Average:", avg)
```

```python
# 6) Dictionary value update + new average
student = {"name": "Rafi", "marks": [70, 80, 90]}
print("Before:", student)

new_mark = 85
student["marks"].append(new_mark)

# manual average
total = 0
count = 0
for m in student["marks"]:
    total += m
    count += 1
avg = total / count if count else 0

print("After:", student)
print("New average:", avg)
```

```python
# 17) math: sqrt, factorial, cube
import math
try:
    n_str = input("Enter a number: ")
    n = float(n_str)
except Exception:
    print("Input not available -> using default n=9")
    n = 9.0

# sqrt (for non-negative)
if n >= 0:
    print("sqrt:", math.sqrt(n))
else:
    print("sqrt: not defined for negative")

# factorial requires integer & non-negative
n_int = int(n)
if n_int >= 0 and abs(n - n_int) < 1e-9:
    print("factorial:", math.factorial(n_int))
else:
    print("factorial: requires a non-negative integer; using int(n) if suitable.")

print("n^3:", n ** 3)
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, value):
        self.items.append(value)
        print("Pushed:", value)

    def pop(self):
        if not self.items:
            print("Stack is empty!")
            return None
        value = self.items.pop()
        print("Popped:", value)
        return value

    def display(self):
        print("Stack (top -> bottom):", list(reversed(self.items)))

# Demo
s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.display()
s.pop()
s.display()
```

```python
# Bonus) Read 'name,marks' lines from file into a dictionary
FILENAME = "students.txt"
# Create a sample file if not present
try:
    open(FILENAME, "r", encoding="utf-8").close()
except FileNotFoundError:
    with open(FILENAME, "w", encoding="utf-8") as f:
        f.write("Rafi,78\nNila,85\nTuhin,90\n")

students = {}
with open(FILENAME, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        if ',' in line:
            name, marks = line.split(',', 1)
            name = name.strip()
            try:
                marks = float(marks.strip())
            except ValueError:
                continue
            students[name] = marks

print("Students dict:", students)
```
